```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Simulate two cointegrated time series
t = np.arange(100)
# Simulate the first series (Stock A)
stock_a = 2 * t + np.random.normal(0, 5, 100)

# Simulate the second series (Index A), which is cointegrated with Stock A
index_a = 3 * t + stock_a + np.random.normal(0, 10, 100)

# Plot the two time series
plt.figure(figsize=(12, 6))
plt.plot(t, stock_a, label='Stock A')
plt.plot(t, index_a, label='Index A')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.title('Simulated Cointegrated Time Series')
plt.grid(True)

# Perform cointegration analysis
result = sm.OLS(stock_a, sm.add_constant(index_a)).fit()
beta, alpha = result.params

# Print the cointegration results
print("Cointegration Test Results:")
print("Alpha (intercept):", alpha)
print("Beta (coefficient of Index A):", beta)
print("H0: Beta = 0 (No cointegration)")
print("H1: Beta < 0 (Cointegration)")

# Display the plot
plt.show()
print(result.summary())
```
